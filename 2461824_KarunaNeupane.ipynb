<h1>Problem - 1A - Single Year HDI Exploration (Latest Year: 2022)</h1>

```python
# Objective:
# Explore the HDI dataset for the latest available year (2022) to practice basic EDA techniques.

# Tasks:
# Complete all the Following Tasks:

# 1. Extract Latest Year:
# • Identify unique years in the dataset.
# • Filter the dataset to include only observations from the year 2022.
# • Save the filtered dataframe as hdi 2022 df (used for all subsequent tasks in Problem 1A).

# Identify unique years in the dataset.
df['year'].unique()
print(f"Unique years in dataset: {sorted(df['year'].unique())}")

# Filter the dataset to include only observations from the year 2022.
hdi_2022_df = df[df['year'] == 2022].copy()
print(f"\n2022 dataset shape: {hdi_2022_df.shape}")
```

```python
# 2. Data Exploration:
# • Display the first 10 rows of the 2022 dataset.
# • Count the number of rows and columns.
# • List all column names and their data types.

print("The first 10 rows of the 2022 dataset:")
hdi_2022_df.head(10)
```

```python
print("The number of rows and columns:")
hdi_2022_df.shape
```

```python
print("List of all column names and their data types:")
hdi_2022_df.info()
```

```python
# 3. Missing Values & Data Cleaning:
# • Check for missing values in each column and report total counts.
# • Inspect dataset for:
# – numeric columns stored as text,
# – inconsistent or misspelled country names,
# – duplicate rows,
# – special characters (e.g., “–”) representing missing data.
# • Apply necessary cleaning steps:
# – convert data types where needed,
# – remove duplicates,
# – handle missing values (drop or impute; justify your choice).

print("The missing values in each column and total counts before cleaning:")
hdi_2022_df.isna().sum()
```

```python
# Inspect dataset for special characters representing missing values

# Replace special characters with NaN
hdi_2022_df.replace("–", np.nan, inplace=True)

# Check for duplicates
duplicates = hdi_2022_df.duplicated().sum()
print(f"Duplicate rows: {duplicates}")

# Convert numeric columns stored as text to proper numeric type
numeric_cols = ['hdi', 'life expectancy', 'gross inc percap', 'gender development']
for col in numeric_cols:
    if col in hdi_2022_df.columns:
        hdi_2022_df[col] = pd.to_numeric(hdi_2022_df[col], errors='coerce')

# Remove duplicate rows
hdi_2022_df.drop_duplicates(inplace=True)

# Drop rows with where HDI (core metric) is missing
hdi_2022_df.dropna(subset=['hdi'], inplace=True)
print(f"After cleaning: \nShape: {hdi_2022_df.shape}")
```

```python
# 4. Basic Statistics:
# • Compute the mean, median, and standard deviation of HDI for the year 2022.
# • Identify the country with the highest HDI in 2022.
# • Identify the country with the lowest HDI in 2022.

print("Mean:")
print(hdi_2022_df['hdi'].mean())

print("\nMedian:")
print(hdi_2022_df['hdi'].median())

print("\nStandard Deviation:")
print(hdi_2022_df['hdi'].std())

print("\nThe country with the highest HDI in 2022:")
print(hdi_2022_df.loc[hdi_2022_df['hdi'].idxmax()][['country','hdi']])

print("\nThe country with the lowest HDI in 2022:")
print(hdi_2022_df.loc[hdi_2022_df['hdi'].idxmin()][['country','hdi']])
```

```python
# 5. Filtering and Sorting:
# • Filter countries with HDI {"hdi"} greater than 0.800.
# • Sort this filtered dataset by Gross National Income (GNI) per Capita {"gross inc percap"} in descending order.
# • Display the top 10 countries.

# Filter countries with HDI {"hdi"} greater than 0.800 and sorting this filtered dataset in descending order.
high_hdi = hdi_2022_df[hdi_2022_df['hdi'] > 0.800]
high_hdi.sort_values(by='gross_inc_percap', ascending=False)

# Display the top 10 countries.
print("\nTop 10 high HDI countries by GNI:")
print(high_hdi[['country', 'hdi', 'gross_inc_percap']].head(10))
```

```python
# 6. Adding HDI Category Column:
# • Create a new column HDI Category that classifies each country into one of the four official Human Development Index groups. 
# The classification should be based on the HDI value for the year 2022. 
# Use the following categories and thresholds defined by the United Nations Development
# Programme (UNDP):
# HDI Category HDI Range (hdi)
# Low < 0.550
# Medium 0.550 – 0.699
# High 0.700 – 0.799
# Very ≥ 0.800
# After creating this new column:
# • verify that all countries are classified correctly,
# • ensure the updated dataframe includes the new category column.
# • Save the final dataframe as HDI category added.csv and include this file in your final submission.

def classify_hdi(hdi):
    if hdi < 0.55:
        return "Low"
    elif hdi < 0.70:
        return "Medium"
    elif hdi < 0.80:
        return "High"
    else:
        return "Very High"

# Verify that all countries are classified correctly.
hdi_2022_df['HDI Category'] = hdi_2022_df['hdi'].apply(classify_hdi)

# Ensure the updated dataframe includes the new category column.
print("\nHDI Category distribution:")
hdi_2022_df['HDI Category'].value_counts()
```

```python
#Saving the final dataframe as HDI_category_added.csv
hdi_2022_df.to_csv("HDI_category_added.csv", index=False)
```

<h1>Problem 2: Advanced HDI Exploration</h1>

```python
# Objective:

# Perform advanced analysis of HDI data, focusing on South Asian countries, composite metrics, 
# outlier detection, metric relationships, and gap analysis.

# Tasks:
# Complete all the following tasks:
# 1. Create South Asia Subset:
# • Define the list of South Asian countries: ["Afghanistan", "Bangladesh", "Bhutan", "India", "Maldives", "Nepal", "Pakistan", "Sri Lanka"].
# • Filter the HDI dataset to include only these countries.
# • Save the filtered dataset as HDI SouthAsia.csv and include this file in the final submission.

# South Asia Subset
south_asia = ["Afghanistan","Bangladesh","Bhutan","India","Maldives","Nepal","Pakistan","Sri Lanka"]

# Filter the HDI dataset to include only these countries and saving the filtered dataset as HDI SouthAsia.csv.
sa_df = df[df['country'].isin(south_asia)].copy()
sa_df.to_csv("HDI_SouthAsia.csv", index=False)

print(f"South Asia dataset: {sa_df.shape}")
```

```python
# 2. Composite Development Score:
# • Create a new metric called Composite Score using the formula:

# Composite Score = 0.30 × Life Expectancy Index + 0.30 × GNI per Capita Index}
# Here: Life Expectancy Index → "life expectancy" and GNI per Capita Index → "gross inc percap"

# Replace special characters and convert to numeric
sa_df.replace("–", np.nan, inplace=True)

# Convert numeric columns stored as text to proper numeric type
sa_df['life_expectancy'] = pd.to_numeric(sa_df['life_expectancy'], errors='coerce')
sa_df['gross_inc_percap'] = pd.to_numeric(sa_df['gross_inc_percap'], errors='coerce')

# Drop rows with missing values needed for composite score
sa_df.dropna(subset=['life_expectancy', 'gross_inc_percap'], inplace=True)

# Create Composite Score
sa_df['Composite Score'] = 0.30 * sa_df['life_expectancy'] + 0.30 * sa_df['gross_inc_percap']
```

```python
# • Rank South Asian countries based on Composite Score.

top5 = sa_df[sa_df['year'] == 2022].sort_values('Composite Score', ascending=False)
print("\nSouth Asia Composite Score Ranking (2022):")
print(top5[['country', 'Composite Score', 'hdi']].head())
```

```python
# • Plot the top 5 countries in a horizontal bar chart.
# • Compare the ranking of countries by Composite Score with their HDI ranking and discuss any differences.


# Horizontal bar chart
plt.figure(figsize=(10, 6))
top5_sa = top5.head(5)
sns.barplot(data=top5_sa, y='country', x='Composite Score')
plt.title('Top 5 South Asian Countries by Composite Score (2022)')
plt.xlabel('Composite Score')
plt.tight_layout()
plt.show()
```

```python
# 3. Outlier Detection:
# • Detect outliers in HDI and GNI per Capita using the 1.5 × IQR rule.

def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    return df[
        (df[column] < Q1 - 1.5 * IQR) |
        (df[column] > Q3 + 1.5 * IQR)
    ]
hdi_outliers = detect_outliers(sa_df, 'hdi')
gni_outliers = detect_outliers(sa_df, 'gross_inc_percap')
```

```python
# • Create a scatter plot of GNI per Capita vs HDI, highlighting the outliers in a different color.

sns.scatterplot(data=sa_df, x='gross_inc_percap', y='hdi', alpha=0.7)

sns.scatterplot(data=gni_outliers, x='gross_inc_percap', y='hdi', color='red', label='Outliers')

plt.title("South Asia: GNI per Capita vs HDI (Outliers Highlighted)")
plt.xlabel("GNI per Capita Index")
plt.ylabel("HDI")
plt.legend()
plt.show()
```

```python
# 4. Exploring Metric Relationships:
# • Select two HDI components (e.g., Gender Development Index {"gender development"} and Life Expectancy Index {"life expectancy"}).
# • Compute Pearson correlation of each metric with HDI.
# • Create scatter plots with trendlines to visualize the relationships.
# • Discuss:
# – Which metric is most strongly related to HDI and shows the weakest relationship with HDI.

# Pearson correlation of each metric with HDI.
sa_df[['hdi','life_expectancy','gender_development']].corr()
```

```python
# 5. Gap Analysis:
# • Create a new metric:

# GNI HDI Gap = "gross inc percap" − "hdi"
# • Rank South Asian countries by GNI HDI Gap in descending and ascending order.
# • Plot the top 3 positive gaps and top 3 negative gaps.
# • Discuss the implications of the gap, e.g., cases where GNI is high but HDI is lower than expected.

sa_df['GNI_HDI_Gap'] = sa_df['gross_inc_percap'] - sa_df['hdi'] * 10000

gap_desc = sa_df.groupby('country')['GNI_HDI_Gap'] \
    .mean().sort_values(ascending=False)

gap_asc = sa_df.groupby('country')['GNI_HDI_Gap'] \
    .mean().sort_values()

top_positive = gap_desc.head(3)
top_negative = gap_asc.head(3)
```

```python
# 7. Outlier Detection:
# • Detect outliers in HDI {"hdi"} and GNI per Capita {"gross inc percap"} for each region using the 1.5 × IQR rule.
# • Create scatter plots highlighting outliers in a different color.
# • Discuss the significance of these outliers

def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

```
